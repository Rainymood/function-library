# Side by side bar chart matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

before = np.array([10, 11, 9, 20])
after = np.array([12, 15, 10, 17])
labels = ["Group " + x for x in list("ABCDEFGHIJLKMNOPQ")[0 : len(before)]]

# Plot
fig, ax = plt.subplots(figsize=(8, 6))
width = 0.4
x_locs = np.arange(len(before))
ax.bar(
    x=x_locs - width, height=before, width=width, color="cornflowerblue", label="Males"
)
ax.bar(x=x_locs, height=after, width=width, color="hotpink", label="Females")

# Pretty up graph
ax.set_xlabel("Gr")
ax.set_xticks(range(len(before)))
ax.set_xticklabels(labels)
ax.yaxis.grid(True)
ax.legend(loc="best")
ax.set_ylabel("Mean group result")
ax.set_title("Group results by gender")
fig.tight_layout(pad=1)
```
