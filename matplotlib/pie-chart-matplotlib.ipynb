```python
import matplotlib.pyplot as plt
import numpy as np
import random

random_style = random.choice(plt.style.available)
plt.style.use(random_style)
print(f"Using style {random_style}")

data = np.array([5, 3, 5, 10])
labels = ["Group " + x for x in list("ABCEFGHIJKLMNOPQ"[0 : len(data)])]
explode = (0, 0.1, 0, 0)  # explode one group
fig, ax = plt.subplots(figsize=(8, 4))
ax.pie(x=data, explode=explode, labels=labels, autopct="%1.1f%%", startangle=270)
plt.show()
```
