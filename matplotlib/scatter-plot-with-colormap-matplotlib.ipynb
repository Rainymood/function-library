```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import random

random_style = random.choice(plt.style.available)
plt.style.use(random_style)
print(f"Using style {random_style}")

N = 100
x = np.random.rand(N)
y = np.random.rand(N)
size = ((np.random.rand(N) + 1) * 8) ** 2
color = np.random.rand(N)

fig, ax = plt.subplots(figsize=(8,6))
plot = ax.scatter(x, y, s=size, c=color, cmap=cm.viridis)

colorbar = fig.colorbar(plot)
colorbar.ax.set_title("color bar title")
ax.grid(True)
fig.tight_layout(pad=1)
```
